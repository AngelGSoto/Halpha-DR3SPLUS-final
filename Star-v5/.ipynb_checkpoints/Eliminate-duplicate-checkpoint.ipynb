```python
import splusdata
import getpass
import pandas as pd
from astropy.table import Table
from pathlib import Path
```

```python
# Connecting with SPLUS database
username = input(prompt="Login: ")
password = getpass.getpass("Password: ")
conn = splusdata.connect(username, password)
```

```python
table = Table.read("Halpha-DR3_PStotal-STAR_total-clean.ecsv", format="ascii.ecsv")
table
```

```python
#Selecting columns
col = ["Field", "ID", "RA", "DEC"]
table_new = table[col]
table_new
```

```python
idd1 = []
for i in table_new:
    if i["ID"].endswith(" '"):
        id1 = i["ID"].split("b'")[-1].split(" ")[0]
        idd1.append(id1)
    else:
        id1 = i["ID"].split("b'")[-1].split("'")[0]
        idd1.append(id1)
```

```python
table_new["ID1"] = idd1
table_new
```

```python
Col_new = ["ID1", "RA", "DEC"]
table_new_ = table_new[Col_new]
table_new_.rename_column('ID1', 'ID')
table_new_
```

```python
Query = f"""SELECT spid.ID, upl.ra, upl.dec 
                FROM TAP_UPLOAD.upload AS upl 
                LEFT OUTER JOIN "idr3"."vac_splusid" AS spid ON upl.id = spid.ID WHERE splusIDflag = 0"""
```

```python
df = (table_new_.to_pandas())
results = conn.query(Query, df)
```

```python
results
```

```python
results.write("Halpha-DR3_PStotal-STAR_total-clean-coor.ecsv", format="ascii.ecsv", overwrite=True)
df_table = results.to_pandas()
df_table.to_csv("Halpha-DR3_PStotal-STAR_total-clean-coor.csv", index = False)
```
